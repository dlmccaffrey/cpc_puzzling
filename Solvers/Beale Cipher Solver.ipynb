# Beale Cipher Solver
A <a href="https://en.wikipedia.org/wiki/Beale_ciphers">Beale cipher</a> is a series of numbers that are word indexes for a specific document, such as the Declaration of Independence. The first letters of the indexed words spell out the cipher.

```python
# first paragraph of the zebra wikipedia page
with open('zebra_wiki.txt') as file:
    text = file.read()
text
```

```python
# remove all the extraneous punctuation and split
word_list = text.replace('\n\n', ' ').replace('(', '').replace(')','').replace(' –', '').split(' ')
word_list
```

```python
for i in range(len(word_list)):
    print(i, word_list[i])
```

```python
cipher = [24, 124, 166, 1, 40, 22, 51, 116]
```

```python
answer = ''
for index in cipher:
    answer += word_list[index-1][0].lower() # Beale ciphers start at 1
answer
```
